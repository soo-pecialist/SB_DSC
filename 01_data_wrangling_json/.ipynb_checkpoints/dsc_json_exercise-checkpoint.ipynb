## imports for Python, Pandas

```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
def print_json(loc, indent=4, sort=True):
    """ To print JSON data in easily readable way
        Warning: 
        File is large. Therefore this function will result in taking up large space"""
    with open(loc) as jsonfile:
        data_list = json.load(jsonfile)
        for data in data_list:
            print(json.dumps(data, indent=indent, sort_keys=sort))
            print()
```

```python
f = 'data/world_bank_projects.json'
```

```python
# print_json(f)
```

```python
df = pd.read_json(f)
df.head()
```

## Q1. Find the 10 countires with most projects

```python
dff = df[['countryname', 'project_name']]
dff.head()
```

```python
def qestuion_1(loc):
    """Input: location of JSON data
       Output: Series of Top 10 countries with the most projects in descending order"""
    
    df = pd.read_json(f)
    df = df[['countryname', 'project_name']]
    
    return df['countryname'].value_counts()[:10]

q1 = qestuion_1(f)
q1
```

```python
def question1_plot(q1):
    """Plot bar graph of top 10 countires with most projects"""
    
    labels = pd.Series(q1.index.str.split()).map(lambda x: x[-1]).tolist()
    color = cm.Paired(np.arange(10))
    ax = q1.plot(kind='bar', rot=30, figsize=(13, 8),
                 fontsize=14, width=.8, alpha=.9, color=color)

    ax.set_xticklabels(labels)
    ax.set_title('Top 10 Countries with the Most Projects', fontsize=15)
    ax.set_xlabel('Countires', fontsize=14)
    ax.set_ylabel('Number of Projects', fontsize=14)

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.grid(color='k', axis='y', linestyle='dashdot', alpha=.2)

question1_plot(q1)
```

## Q2. Find the top 10 major project themes

```python
dff = df['mjtheme_namecode']
dff.head()
```

```python
def question_2(loc):
    """Input: location of JSON file
       Output: Series of top 10 major project themes in descending order"""
    
    def extract_name(element):
        if element == -1:
            return None
        else:
            return element['name']
    
    df = pd.read_json(f)
    # list of dict
    dff = df['mjtheme_namecode'] 
    # each entry of list to columns
    dff2 = pd.DataFrame(dff.values.tolist()) 
    # filling in -1 to process further
    dff2 = dff2.fillna(-1)
    # extract project theme: dict['name']
    dff2 = dff2.applymap(extract_name)
    # empty Series to concatenate all values
    df = pd.Series()
    
    for i in range(len(dff2.columns)):
        df = pd.concat([df, dff2[i]], axis=0, ignore_index=True)
    
    return df.value_counts()[:10]

q2 = question_2(f)
q2
```

```python
def question2_plot(q2):
    """Plot bar graph of top 10 project themes"""
    
    def multiline_index(labels):
        new_labels = []
        for entry in labels:
            if len(entry) > 3:
                entry.insert(3, '\n')
            label = ' '.join(entry)
            new_labels.append(label)
        return new_labels
    
    temp = pd.Series(q2.index).str.split().values.tolist()
    labels = multiline_index(temp)
    
    color = cm.Paired(np.arange(10))
    ax = q2.plot(kind='bar', rot=65, figsize=(13, 8),
                 fontsize=14, width=.8, alpha=.9, color=color)
    ax.set_xticklabels(labels)
    ax.set_title('Top 10 Project Themes', fontsize=15)
    ax.set_xlabel('Project Themes', fontsize=14)
    ax.set_ylabel('Counts', fontsize=14)

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.grid(color='k', axis='y', linestyle='dashdot', alpha=.2)

question2_plot(q2)
```

## Q3. In Q2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in

```python
def question_3(loc):
    """Input: location of JSON data
       Output: Dataframe of project code and theme with missing names filled in"""
    
    with open(loc) as json_file:
        data = json.load(json_file)
        df = json_normalize(data, 'mjtheme_namecode')
        df['name'] = df['name'].replace('', '**Missing Project Theme**')

    return df

q3 = question_3(f)
```
